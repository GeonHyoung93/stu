## 결측치(missing data) 처리하기 
* null, NaN, NA, None
* 파이썬/pandas에서는 NaN(float, 숫자)또는 None(object, 문자)으로 취급
* 단, 정수형 누락값인 NA는 pandas에서는 취급 불가
* 결측치는 numpy 모듈을 이용해서 정의



```python
a = np.array([1, 2, None, 4, 5])
print(a, a.dtype)
print(a * 100)
print(a.sum())
```

```python
# numpy 에서 제공하는 결측치 상수로 데이터 정의
b = np.array([1,2,np.NaN,4,5])
print(b, b.dtype)
print(b * 100) # 결측치 제외 연산결과 출력 
print(b.sum()) # 결측치 때문에 연산 불가
```

```python
# pandas 객체로 정의한 뒤 결측치 처리
# c = pd.Series([1,2,np.NaN, 4,5])
c = pd.Series([1,2,None, 4,5])
print(c, c.dtype)
print(c * 100)
print(c.sum())
```

## 결측치 처리
* pandas 자료구조에서는 null 값을 감지하고 삭제하는 기능 제공
    + isnull, dropna, notnull, fillna

```python
d = pd.Series([1,2,np.NaN, 4, None])
# d.isnull() # NaN, None 여부 출력
# d.isnull().value_counts()  # NaN, None 여부 False/True로 카운트 해서 출력
# d.notnull().value_counts()

d[d.isnull()]  # 결측치가 있는 인덱스 출력
```

```python
d.dropna() # 결측치 제거, 원본유지
```

```python
d.fillna(0)  # 결측치 대체
```

## 사원데이터에 존재하는 결측치 처리
* 수당의 결측치는 0.0으로 설정
